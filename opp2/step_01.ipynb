```python
class product :
    def __init__(self,name,price,quantity):
        self.__name = name
        self.__price = price
        self.__quantity = quantity
        
    @property  
    def name(self):
        return f"{self.__name}"
    
    @property     
    def price(self):
        return self.__price
    
    @property 
    def quantity(self):
        return self.__quantity
    
    @name.setter    
    def name(self,name):
        self.__name = name
        return f"güncel isim: {self.__name}"
    
    @price.setter
    def price(self,price):
        self.__price = price
        return f"güncel isim: {self.__price}"
    
    @quantity.setter
    def quantity(self,quantity):
        self.__quantity = quantity
        return f"güncel isim: {self.__quantity}"
    

    def get_total_price(self):
        if isinstance(self.price,(int,float)) and isinstance(self.quantity,(int,float)):
            return f"{self.price * self.quantity} total ücret"
        else:
            return f"toplanamayacak kadar değerli"

    def kayitSilme(self):
        if not self.name == "sefa":
            print(f"{self.__name} isimli kayıt silindi")
            del self.__name
            del self.__price
            del self.__quantity
        else :
            print(f"{self.name} isimli kaydı silemezsin layn")
        
```

```python
item1 = product("sefa","paha biçilmez","eşsiz")

print(item1.name)
print(item1.price)
print(item1.quantity)
print(item1.get_total_price())
item1.kayitSilme()
```
